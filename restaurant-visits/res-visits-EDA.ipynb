```python
import pandas as pd
```

```python
air_store_info = pd.read_csv('./input_data/air_store_info.csv')
hpg_store_info = pd.read_csv('./input_data/hpg_store_info.csv')
store_id_relation = pd.read_csv('./input_data/store_id_relation.csv')
```

```python
store_id_relation.merge(air_store_info, on = 'air_store_id').shape
```

```python
df = store_id_relation.merge(air_store_info, on='air_store_id', how = 'left')
df = df.merge(hpg_store_info, on='hpg_store_id', suffixes =('_air', '_hpg'), how = 'inner')
df
```

```python
import distance as di
import numpy as np

def coord_distance(lat1, long1, lat2, long2):
    return np.sqrt( (lat2 - lat1)**2 + (long2 - long1)**2 )

def string_distance(desc1, desc2):
    return di.jaccard(desc1, desc2)

df['coord_distance'] = coord_distance(df['latitude_air'], df['longitude_air'], df['latitude_hpg'], df['longitude_hpg'])
df['area_distance']= df.apply(lambda row: string_distance(row['air_area_name'], row['hpg_area_name']), axis = 1)
df['genre_distance']= df.apply(lambda row: string_distance(row['air_genre_name'], row['hpg_genre_name']), axis = 1)
df
```
